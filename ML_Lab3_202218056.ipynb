```python
def get_points(n, std):
  xvalues = np.random.uniform(size=n)
  yvaluese = list()
  yvalues = list()
  e = np.random.normal(0, std, n)
  for i in range(n):
    yvalues.append(np.sin(2*math.pi*xvalues[i]))
    yvaluese.append(np.sin(2*np.pi*xvalues[i])+e[i])
  return xvalues, yvaluese, yvalues
```

```python
xpoints, ypoints, actualY=get_points(500, 0.25)
```

```python
def get_coeff(deg, xval, yval):
  mat=[]
  rows=len(xval)
  for i in range(rows):
    a=[]
    for j in range(deg+1):
      a.append(xval[i]**(deg-j))
    mat.append(a)
  matT = np.transpose(mat)
  inv = np.linalg.inv(np.dot(matT,mat))
  final = np.dot(np.dot(inv,matT),yval)
  return final
```

```python
def fun(xval, coeff, deg):
  exp=0
  for i in range(deg+1):
    exp+=coeff[i]*(xval**(deg-i))
  return exp
```

```python
def plot_fun(deg, coeff, xval, yval):
  s=0
  e=1
  x = np.linspace(s,e)
  y = fun(x, coeff, deg)
  plt.plot(xval, yval, '*', label='Data points')
  plt.plot(x,y, label='Fited curve', color='black')
```

```python
def get_rmse(deg, predy, yval):
  mse = np.square(np.subtract(yval,predy)).mean()
  rmse = np.sqrt(mse)
  print("Degree of polynomial : ", deg)
  print("RMSE : ", rmse)
```

```python
def get_coeff1(A, yval, lam):
  AT = np.transpose(A)
  mat=np.dot(AT,A)
  for i in range(len(A[0])):
    mat[i][i]+=lam
  inv = np.linalg.inv(mat)
  coeff = np.dot(inv,AT)
  res = np.dot(coeff,yval)
  return res
```

```python
def dataset(n, coeff):
  xpoints=list()
  ypoints=list()
  zpoints=list()
  xdata = np.linspace(0,1,n)

  ydata = np.linspace(0,1,n)

  for i in range(n) :
    for j in ydata : 
      ypoints.append(j)

  for i in xdata :
    for j in range(n) : 
      xpoints.append(i)
    
  for i in range(n*n):
    zpoints.append(f(xpoints[i],ypoints[i],coeff))

  return xpoints,ypoints,zpoints
```

```python
def plot_points(xval):
  xpoints = list()
  ypoints = list()
  zpoints = list()

  for i in xval : 
    xpoints.append(i[0])
    ypoints.append(i[1])
    zpoints.append(np.sin((2*np.pi*(xpoints[-1]**2 + ypoints[-1]**2)**(0.5))))
    
  return xpoints,ypoints,zpoints
```

```python
def plot_fun1(coeff, xval, yval, n):
  x,y,z = dataset(n,coeff)
  fig = plt.figure(figsize = (5, 5))
  ax = plt.axes(projection = '3d')
  ax.plot3D(x, y, z, 'skyblue')

  x1 = list()
  y1 = list()
  z1 = yval

  for value in xval : 
    x1.append(value[0])
    y1.append(value[1])

  plt.plot(x1,y1,z1,'*' , color = 'r')
  x2,y2,z2 = plot_points(xval)
  plt.legend([ 'Predicted Function' , 'Data points'])
  plt.show()

  fig = plt.figure(figsize = (5, 5))
  ax = plt.axes(projection = '3d')
  ax.plot3D(x, y, z, 'skyblue')
  plt.plot(x2,y2,z2 , '*' , color = 'r')
  plt.legend([ 'Predicted Function' , 'E[y/x]'])
  plt.show()
```

```python
def get_coeff_sigmoidal(deg, xval, yval, lam):
  identity = np.identity(deg+1)
  identity[deg][deg] = 0
  A = np.zeros((n, deg+1))
  for i in range(n):
    for j in range(deg+1):
      A[i][j] = xval[i]**j

  b = np.linspace(0, 1, deg)
  w = np.linspace(0, 1, deg)

  for i in range(n):
    A[i][deg] = 1

  for i in range(n):
    for j in range(deg):
        A[i][j] = (1/(1 + math.exp(-1*(w[j]*xval[i] + b[j]))))

  lam_mat = identity*lam
  AT=np.transpose(A)
  mat=np.linalg.inv(np.dot(AT, A) + lam_mat)
  coeff = np.dot(mat, np.dot(AT, yval))

  predy = []
  rmse=0
  for i in range(n):
    res=0
    for j in range(deg+1):
      res += A[i][j]*coeff[j]
    rmse += (yval[i] - res)**2
    predy.append(res)
  rmse = np.sqrt((rmse)/n)

  return coeff, predy, rmse, w, b
```

```python
def get_kernel(lam, xval, yval, sigma):
  m=len(xval)
  identity = np.identity(m+1)
  identity[m][m] = 0
  A = np.zeros((m, m+1))

  for i in range(m):
      A[i][m] = 1
  for i in range(m):
    for j in range(m):
        A[i][j] = math.exp(-1*(((xval[i] - xval[j])**2)/(2*(sigma**2)))) 

  lam_mat = identity*lam
  AT=np.transpose(A)
  mat=np.linalg.inv(np.dot(AT, A) + lam_mat)
  coeff = np.dot(mat, np.dot(AT, yval))

  predy = []
  rmse=0
  for i in range(n):
    res=0
    for j in range(m+1):
      res += A[i][j]*coeff[j]
    rmse += (yval[i] - res)**2
    predy.append(res)
  rmse = np.sqrt((rmse)/n)

  return coeff, predy, rmse
```
