```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
def getdata(mu , std , len) :
    X = np.random.uniform(size=len)
    E = np .random.normal(mu,std,len)
    Y = list()
    for i in range(len) :
        Y.append(math.sin((44*X[i])/7) + E[i])
    return X , Y
```

```python
X1 , Y1 = getdata(1,0.25,20)
print("x values : ",X1)
print("\ny values : ",Y1)
```

```python
X2 , Y2 = getdata(1,0.25,50)
print("x values : ",X2)
print("\ny values : ",Y2)
```

```python
def RMSE(d , c , actualX , actualY)  :
    
    predicted = list()

    predicted = fun(actualX,d,c)

    MSE = np.square(np.subtract(actualY,predicted)).mean() 
    

    RMSE = math.sqrt(MSE)

    print("\npolynomial of degree " , d)

    print("RMSE : " , RMSE)
```

```python
def get_coefficent(n , m , X , Y) :
    mat = np.ones((n,m+1))
    for i in range(n):
        for j in range(m):
            mat[i][j] = (X[i]**(m-j))
    matT = np.transpose(mat)
    val = np.dot(np.linalg.inv(np.dot(matT,mat)),matT)
    final = np.dot(val,Y)
    return final
```

```python
s = 0
e = 1

def fun(x,deg,co):
    res = 0
    for i in range(deg+1):
        res += (co[i]*(x**(deg-i)))
    return res

def plot_graph(deg , co , X , Y):
    x = np.linspace(s,e)
    y = fun(x,deg,co)
    plt.plot(X,Y,'*')
    plt.plot(x,y)
    plt.legend(['Data points' , 'Fited curve' ])
    plt.grid()
    plt.show()
```

# **Training Model**

```python
x = X1
y = list();
for value in X1 : 
    y.append(math.sin((44*value)/7))
```

Polynomial of order 3

```python
co3 = get_coefficent(len,3,X1 , Y1)
print("Coefficents : " , co3)

plot_graph(3 , co3 , X1 , Y1)
plt.scatter(x, y, s=50, facecolors='none', edgecolors='black')
plt.legend('E[x/y] = sin(2*pi*x)')
```

```python
RMSE(1,co1 , X1 , Y1)
RMSE(2,co2 , X1 , Y1)
RMSE(3,co3 , X1 , Y1)
RMSE(9,co4 , X1 , Y1)
```
